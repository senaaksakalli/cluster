```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import KElbowVisualizer
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df = pd.read_excel('cars.xls')
```

```python
df.head()
```

```python
x = df[['Price', 'Mileage', 'Cylinder']]
```

```python
x.head()
```

```python
# HATA DÜZELTMESİ: n_clusters parametresi açıkça belirtildi, n_init uyarısını gidermek için eklendi
model = KMeans(n_clusters=4, n_init=10, random_state=42)
model = model.fit(x)
tahmin = model.predict(x)   # HATA DÜZELTMESİ: tahmin değişkeni tanımlanmadığı için eklendi
x = x.copy()                # SettingWithCopyWarning'i önlemek için
x['cluster'] = tahmin
```

```python
x.head()
```

```python
print('Silhouette Skoru:', silhouette_score(x[['Price','Mileage','Cylinder']], tahmin))
```

```python
# HATA DÜZELTMESİ: 'fissize' -> 'figsize'
plt.figure(figsize=(10, 6))
plt.scatter(x.Mileage, x.Price, c=tahmin)
plt.colorbar()
plt.xlabel('km')
plt.ylabel('fiyat')
plt.title('KMeans Kümeleme: Mileage vs Price')
plt.show()
```

```python
# HATA DÜZELTMESİ: for satırına eksik ':' eklendi, girintiler düzeltildi
wcss = []
ss = []
for i in range(2, 10):
    model = KMeans(n_clusters=i, n_init=10, random_state=42)
    model = model.fit(x[['Price','Mileage','Cylinder']])
    tahmin = model.predict(x[['Price','Mileage','Cylinder']])
    ss1 = silhouette_score(x[['Price','Mileage','Cylinder']], tahmin)
    ss.append(ss1)
    print(f'K={i}  Silhouette: {ss1:.4f}')
    wcss.append(model.inertia_)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(range(2, 10), wcss)
plt.xlabel('K')
plt.ylabel('WCSS')
plt.title('Elbow Yöntemi')
plt.show()
```

```python
sns.lineplot(x=range(2, 10), y=wcss, marker='*')
plt.title('WCSS - Seaborn')
plt.show()
```

```python
# HATA DÜZELTMESİ: 'K[]bowVisualizer' -> 'KElbowVisualizer'
model = KMeans(random_state=42, n_init=10)

visualizer = KElbowVisualizer(model, k=(2, 10), metric='distortion', timings=False)

visualizer.fit(x[['Price','Mileage','Cylinder']])
visualizer.show()
```

```python
data = linkage(x[['Price','Mileage','Cylinder']], method='ward', metric='euclidean')
```

```python
# HATA DÜZELTMESİ: 'figisze' -> 'figsize'
plt.figure(figsize=(25, 6))
dendrogram(data)
plt.title('Hiyerarşik Kümeleme Dendrogramı')
plt.show()
```
